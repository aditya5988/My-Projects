```python
# Task 2
HR_salary=np.random.randint(25000,50001,500)
print("HR salary is :", HR_salary)
Finance_salary=HR_salary*1.25
print("Finance Salary is :", Finance_salary)
Sales_salary=Finance_salary*0.5
print("Finance salary is: ", Sales_salary)
IT_salary=HR_salary+5000
print("IT Salary is :", IT_salary)
```

```python
# Task 3
Total_HR=np.sum(HR_salary)
Total_Finance=np.sum(Finance_salary)
Total_IT=np.sum(IT_salary)
Total_sales=np.sum(Sales_salary)
print("Total HR :",Total_HR )
print("Total_Finace :", Total_Finance)
print("Total_IT :", Total_IT)
print("Total_sales :", Total_sales)
```

```python
# Task 4
tsal_HR=np.sort(HR_salary)[-1:-11:-1]
print("Top 10 salaries of HR Department :", tsal_HR)
print("The sum of top 10 salaries of HR Department :", tsal_HR.sum())

tsal_Fin=np.sort(Finance_salary)[-1:-11:-1]
print("Top 10 salaries of Finance Department :", tsal_Fin)
print("The sum of top 10 salaries of Finance Department :", tsal_Fin.sum())

tsal_Sales=np.sort(Sales_salary)[-1:-11:-1]
print("Top 10 salaries of Sales Department :", tsal_Sales)
print("The sum of top 10 salaries of Sales Department :", tsal_Sales.sum())

tsal_IT=np.sort(IT_salary)[-1:-11:-1]
print("Top 10 salaries of IT Department :", tsal_IT)
print("The sum of top 10 salaries of It Department :", tsal_IT.sum())




```

```python
# Task 5
HR_sal=np.count_nonzero(HR_salary>45000)
Fin_sal=np.count_nonzero(Finance_salary>45000)
IT_sal=np.count_nonzero(IT_salary>45000)
Sal_sal=np.count_nonzero(Sales_salary>45000)
Tot_emp=HR_sal+Fin_sal+IT_sal+Sal_sal
print("Total employees with salary more than 45000 that are eligible for 10% increment are :", Tot_emp)
```
