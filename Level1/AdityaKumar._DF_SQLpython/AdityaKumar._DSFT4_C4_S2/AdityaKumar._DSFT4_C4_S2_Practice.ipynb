```python
import numpy as np
```

```python
from numpy import genfromtxt
df=genfromtxt("DS1_C4_S2_Admission_Predict_Practice.csv", delimiter=',',skip_header=1)
print(df)
cols={0:'SerialNo',1:'GRE',2:'TOEFEL',3:'Univ Rating',4:'SOP',5:'LOR',6:'CGPA',7:'Research',8:'Chace of Adimission'}
```

```python
# Task 1
ind=np.where(df[:,1]>320)
a=np.count_nonzero(ind)
print(a)
```

```python
# Task 2
ind=np.where(df[:,2]>110)
print("The admission possibilities of students with TOEFEL score greater than 110 are :", df[ind,8])
```

```python
# Task 3
sort_gre=np.sort(df[:,1])[::-1]
print(sort_gre[0:11])
ind_gre=np.where(df[:,1]>=339)
print("The list of top 11 GRE scorers are:", df[ind_gre])

sort_toefel=np.sort(df[:,2])[::-1]
print(sort_toefel[0:11])
ind_toefel=np.where(df[:,2]==120)
print("The list of top 8 TOEFEL scorers are :",df[ind_toefel])
```

```python
# Task 4
ind=np.where(((df[:,1]>299)&(df[:,1]<306))&(df[:,4]==3.5)&(df[:,7]==1))
df1=df[ind]
print("The serial number and toefel scores of these students are: ", df1[0:3:2].T)

```

```python
# Task 5
scholarship=np.random.randint(5000,10000,size=(400,1))
df=np.append(df,scholarship,axis=1)
print(df)
sch=np.where(df[:,9]>9000)
a=np.count_nonzero(sch)
print("Total Students with scholarship more than 9000 are :", a)
```

```python
# Task 6
print("The SOP and LOR scores of students with scholarship more than $9000\n",df[:,[4,5]])



```

```python
# Task 7
sch_1=np.where((df[:,1]>315)&(df[:,2]>115)&(df[:,7]==1))
print("Serial numbers of students eligible for additional scholarship of $1000 are,\n",df[sch_1][:,0])
print("Number of students eligible for additional scholarship =",df[sch_1].shape[0])
```
