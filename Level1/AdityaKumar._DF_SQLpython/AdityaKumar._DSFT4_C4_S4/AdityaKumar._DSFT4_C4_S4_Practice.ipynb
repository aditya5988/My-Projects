```python
# Task 10
urban=df[(df.Property_Area=="Urban")]
semi=df[(df.Property_Area=="Semiurban")]
print("Number of customers with property in Urban area :", len(urban))
print("Number of customers with property in Semiurban area :", len(semi))
diff=abs(len(urban)-len(semi))
print("Difference in number of such customers: ",diff)
peru=(diff/len(urban))*100
print("Number of customers having property in Semiurban area is ",peru,"% more than urban area")
```
