```python
#  Task 2
bmw_2010=df[(df.make=='BMW')&(df.year==2010)&(df.condition>4.5)]
bmw_mean=bmw_2010.groupby(by=['make'])['sellingprice'].mean()
bmw_mean
```

```python
bmw_2015=df[(df.make=='BMW')&(df.year==2015)&(df.condition>4.5)]
bmw_mean15=bmw_2015.groupby(by=['make'])['sellingprice'].mean()
bmw_mean15
```

```python
perc=((bmw_mean15-bmw_mean)/bmw_mean15)*100

print("BMW car in 2015 has a selling price of 45.1741 % more than selling price in 2010", perc)
```
