```python
import pandas as pd
```

```python
df=pd.read_csv('housing.csv')
df
```

```python
# Task 1
df1=df[(df.LotArea>=6000)&(df.LotConfig=='Corner')]
df1
```

```python
# Task 2
df2=df[(df.HouseStyle=='1Story')&(df.BldgType=='1Fam')]
df2
```

```python
# Task 3
df3=df[(df.BldgType=='TwnhsE')]
df3
```

```python
# Task 4
df4=df[(df.BedroomAbvGr>3)&(df.Exterior1st=='CemntBd')&(df.BsmtExposure!='NA')]
df4
```

```python
# Task 5
df5=df[(df.GarageCars==2)&(df.GarageArea>=500)]
df5
```

```python
# Combining the dataframes

df_com=pd.concat([df1,df2,df3,df4,df5],axis=0, ignore_index=True)
df_com
```
