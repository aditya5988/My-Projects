```python
# Conversion app

def print_menu():
    print ('1.   Radians to Degree')
    print ('2.   Number to its Factorial')
    print ('3.   Megabyte to Nibble')
    print ('4.   Km/hr to m/sec')
    print ('5.   Integer to Float')
    print ('6.   Decimal to Octal')
    print ('7.   Decimal to Hexadecimal')
    print ('8.   Decimal to Binary Conversion')
    print ('9.   Character to ASCII')
    print ('10.  Celcius to Fahrenheit')
    print ('11.  Degree to Radian')
    print ('12.  Number to its Square Root')
    print ('13.  Fraction to Decimal')
    print ('14.  m/sec to Km/hr')
    print ('15.  Float to Integer')
    print ('16.  Octal to Deciaml')
    print ('17.  Hexadecimal to Decimal')
    print ('18.  Binary to Decimal')
    print ('19.  ASCII to Character')
    print ('20.  Fahrenheit to Celcius')
    
def rad_deg():
    rad=float(input("Enter a input in radiant :"))
    deg=rad*(180/3.14)
    print("Given Radiant in degrees is : ", deg)
    
def deg_rad(): 
    deg=float(input("Enter a input in Degree :"))
    rad=deg*(3.14/180)
    print("Given Degrees in Radiant is : ", rad)
    
def meg_nibb():
    meg=float(input("Enter the Megabye : "))
    nibb=meg*2000000
    print("Megabyte to nibble is : ", nibb)
    
def km_ms():
    km=float(input("Enter the Value in Kilometer per Hour: "))
    ms=km*(5/18)
    print("Kilometer per Hour into Meter per second is: ", ms)
    
def ms_km():
    ms=float(input("Enter the Value in Meter per Second: "))
    km=ms*(18/5)
    print("Meter per second into Kilometer per Hour is: ", km)
    
def inte_floa():
    inte=int(input("Enter the integer value :"))
    floa=float(inte)
    print("The provided integer value in float is", floa)
    
def floa_inte():
    floa=float(input("Enter the float value :"))
    inte=int(floa)
    print("The provided float value in integer is", inte)
    
def dec_oct():
    dec=int(input("Provide the input : "))
    print(oct(dec), "in octal")
    
def dec_hexa():
    dec=int(input("Provide the input : "))
    print(hex(dec), "in hexadecimal")
    
def dec_bin():
    dec=int(input("Provide the input :"))
    print(bin(dec), "in binary")
    
def ch_asc():
    ch=input("Enter the character :")
    print("ASCII value of ", ch, "is", ord(ch))
    
def asc_ch():
    asc=int(input("Enter the ASCII value :"))
    print("Character of ", asc, "is", chr(asc))
    
def cel_fah():
    cel=float(input("Temperature value in degree celcius: "))
    fah=(cel*(9/5)+32)
    print("The teamperature in fahrenheit is :", fah)
    
def fah_cel():
    fah=float(input("Enter the temperature in Fahrenhite:"))
    cel=(fah-32)*(5/9)
    print("The temperature in degree celcius is : ", cel)
    
def num_sq():
    num=int(input("Enter the number :"))
    sq=num**0.5
    print("The square root of ", num,"is", sq)
    
def frc_dec():
    num1=int(input("Enter the first number"))
    num2=int(input("Enter the second number"))
    frc=num1/num2
    dec=float(frc)
    print("The decimal values ff given fraction is :", dec)
    
def num_frac():
    from fractions import Fraction
    num1=int(input("Enter the first number"))
    num2=int(input("Enter the second number"))
    frc=Fraction(num1,num2)
    print(frc)
    
def oct_dec():
    oct = input("Enter an octal number :")
    dec = int(oct,8)
    print("The decimal value of {} is {}".format(oct,dec))
    
def hex_dec():
    hex = input("Enter the value")
    dec = int(hex, 16)
    print('Value in decimal :', dec)
    
def bin_dec():
    bin= input('Enter a Binary number:')
    dec= int(bin, 2)
    print('The decimal conversion is:', dec)
    
    
    
def fact_num():
    n=int(input("Enter a number: "))
    num1=1
    
    
    while(n>=2):
        num1=num1*n
        n=n-1
    print("Factorial of given number is: ", num1)
    

    

    
    

    
    
    
    


print_menu()

choice = input("Which conversion would you like to do? : ")
if choice=='1':
    rad_deg()
if choice=='2':
    fact_num()
    
                  
if choice=='3':
    meg_nibb()
    
if choice=='4':
    km_ms()
    
if choice=='5':
    inte_floa()
    
if choice=='6':
    dec_oct()
    
if choice=='7':
    dec_hexa()
    
if choice=='8':
    dec_bin()
    
if choice=='9':
    ch_asc()
    
if choice=='10':
    cel_fah()
    
if choice=='11':
    deg_rad()
     
    
if choice=='12':
    num_sq()
    
if choice=='13':
    frc_dec()
    
if choice=='14':
    ms_km()
    
if choice=='15':
    floa_inte()
    
if choice=='16':
    oct_dec()
    
if choice=='17':
    hex_dec()
    
if choice=='18':
    bin_dec()
    
if choice=='19':
    asc_ch()
    
if choice=='20':
    fah_cel()
    
    
    
```
