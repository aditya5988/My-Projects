```python
stock={1:['Rice',15,112.80],2:['Dal',10,223.40],3:['Butter',5,46.50],4:['Cheese',0,69.50],5:['Bread',20,21.20],
      6:['Egg',60,5.40],7:['Chicken',3,200.50],8:['Peas',7,34.70],9:['oil',20,175.30],10:['Water',40,28.70]}


# menu
print("Welcome to Store")
print("Please place your Order")
print()
print('%4s %10s %10s %10s'%('S.no','Item','Stock_Qty','Cost/item'))
for key,value in stock.items():
    print('%4d %10s %6d %15f' %(key,value[0],value[1],value[2]))
    
# Placing orders
print()
print("If u don't want an item please enter 0.".upper())
purchase= {}
for key,value in stock.items():
    print("Enter the quantity of %s needed: " % value[0])
    a = int(input())
    while a> value[1]:
        print("Sorry! No stock available.")
        print("Enter the quantity of %s needed:"% value[0])
        a=int(input())
    purchase.update({key:a})
#print(purchase)

# Details
print()
print()
print("Please provide address for Home Delivery")
print()
name=input("Enter your name: ")
address= input("Enter your address: ")
print()
print("No delivery for less than 5 kms")
distance= int(input("How far do you live from our Shop in kms:"))
if distance<=10:
    dist_amt=0
elif distance<=15:
    dist_amt=25
elif distance>15:
    dist_amt=50
    
# Cost calculation ##
cost ={}
for i in stock:
    cost.update({i:[stock[i][0],purchase[i],stock[i][2]*purchase[i]]})
    
# Calculatig the bill

print()
print("%22s"% "Bill")
print('%4s %10s %10s %8s'%('S.no', 'Item', 'Purchase_qty', 'Cost'))
total_cost = 0
for key, value in cost.items():
    print('%4d %10s %6d %16f' %(key,value[0],value[1],value[2]))
    total_cost+=value[2]
    
print()

# Printing cost
print("Total cost = ", total_cost)
gst=(total_cost*0.12)
print("GST is 12% =", gst)
print("Delivery charge =", dist_amt)
billing_amount= total_cost+gst+dist_amt
print("Total Billing Amount is = ", billing_amount)
print()

# Address and Other details

print("%s" % "Billing Address")
print("Name :", name)
print("Address: ", address)

# Goodbye
print()
print("Thanks for choosing us. Hope to see you again")


```
