```python
cars = cars1['Make'].value_counts()
print(cars.index)
 
for i in cars.index:
    cars3 = cars1[cars1['Make'] == i]
    print (i)
        
```

```python
cars = cars1['Make'].value_counts()
print(cars.index)
 
for i in cars.index:
    cars3 = cars1[cars1['Make'] == i]
    print (i)
    Mean = []
    Median = []
    Std = []
    Covar = []

    cols = ['Displacement', 'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 'Highway_Mileage_km_litre', 'Minimum_Turning_Radius']
    for i in cols:
        Mean.append(cars3[i].mean())
        Median.append(cars3[i].median())
        Std.append(cars3[i].std())
        Covar.append((cars3[i].mean()/cars1[i].std())*100)

    row_names = ['Mean', 'Median', 'Std', 'Covar']

    cars_meas = pd.DataFrame([Mean, Median, Std, Covar], index = row_names, columns = cols)
    print(cars_meas)
```
