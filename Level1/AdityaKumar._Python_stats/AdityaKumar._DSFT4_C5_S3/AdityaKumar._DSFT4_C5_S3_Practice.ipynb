```python
# Task 3
import pandas as pd
df=pd.read_csv('DS1_C5_S3_Sampling_Practice_GamesSales_Data.csv')
df
```

```python
# Task 3.1.1
# Random Sampling without replacement
game_df=df.sample(n=30, random_state=1)
game_df
```

```python
# Task 3.1.2
# Random Sampling with replacement
game_df1=df.sample(n=30, replace=True , random_state=1)
game_df1
```

```python
# Task 3.1.3
# Systemic sampling
import numpy as np
indexes=np.arange(0,len(df),step=50)
df.iloc[indexes]
```

```python
# Task 3.1.4
# Strafied sampling
grouped_d_df=df.groupby(['Year'])['Genre'].value_counts()
grp_df=df.groupby(['Year','Genre'])
Strat_Sample_df=pd.DataFrame()
for y in grouped_d_df.index:
    d_data=grp_df.get_group(y)
    sample_data=d_data.sample(n=30,replace=True,random_state=1)
    Strat_Sample_df=Strat_Sample_df.append(sample_data, ignore_index=True)
    
Strat_Sample_df
```

```python
# Task 3.1.5
# Clustered sampling
grouped_year_df=df.groupby(['Year'])
years=[2016, 2019]
C_sample_df=pd.DataFrame()
for y in years:
    year_data=grouped_year_df.get_group(y)
    sample_data=year_data.sample(frac=.1 , random_state=1)
    C_sample_df=C_sample_df.append(sample_data, ignore_index=True)
    
C_sample_df
```

```python
grouped_genre_df1=df.groupby(['Genre'])
genres=['Action','Adventure']
C_sample_df1=pd.DataFrame()
for g in genres:
    genre_data=grouped_genre_df1.get_group(g)
    sample_data=genre_data.sample(frac=.1 , random_state=1)
    C_sample_df1=C_sample_df.append(sample_data, ignore_index=True)
    
C_sample_df1
```

```python
# Task 3.2.1

sample_data10=df.sample(n=10, replace=True, random_state=1)
sample_data20=df.sample(n=20, replace=True, random_state=1)
sample_data30=df.sample(n=30, replace=True, random_state=1)
print("Game data with sample size 10 ",sample_data10)
print("Game data with sample size 20 ",sample_data20)
print("Game data with sample size 30 ",sample_data30)
```

```python
import statistics as st
import scipy.stats as sts
```

```python
sample_mean10=st.mean(sample_data10['Global_Sales_M$'])
sample_sd10=st.stdev(sample_data10['Global_Sales_M$'])
sample_median10=st.median(sample_data10['Global_Sales_M$'])
print("Mean of game data with 10 sample space =",sample_mean10)
print("Sd of game data with 10 sample space =",sample_sd10)
print("Median of game data with 10 sample space = ",sample_median10)

sample_mean20=st.mean(sample_data20['Global_Sales_M$'])
sample_sd20=st.stdev(sample_data20['Global_Sales_M$'])
sample_median20=st.median(sample_data20['Global_Sales_M$'])
print("Mean of game data with 20 sample space =",sample_mean20)
print("Sd of game data with 20 sample space =",sample_sd20)
print("Median of game data with 20 sample space = ",sample_median20)

sample_mean30=st.mean(sample_data30['Global_Sales_M$'])
sample_sd30=st.stdev(sample_data30['Global_Sales_M$'])
sample_median30=st.median(sample_data30['Global_Sales_M$'])
print("Mean of game data with 30 sample space =",sample_mean30)
print("Sd of game data with 30 sample space =",sample_sd30)
print("Median of game data with 30 sample space = ",sample_median30)




```

```python
# Task 3.2.2
sample_mean_pop=st.mean(game_df['Global_Sales_M$'])
sample_sd_pop=st.stdev(game_df['Global_Sales_M$'])
sample_median_pop=st.median(game_df['Global_Sales_M$'])
print("Mean of game data with population data =",sample_mean_pop)
print("Sd of game data with population data =",sample_sd_pop)
print("Median of game data with population data = ",sample_median_pop)
```
