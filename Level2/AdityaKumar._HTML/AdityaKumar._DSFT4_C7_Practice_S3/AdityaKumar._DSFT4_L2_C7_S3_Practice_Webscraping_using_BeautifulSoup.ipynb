```python
import sys
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Task 1

url='https://en.wikipedia.org/wiki/Leonardo_DiCaprio_filmography'

response=requests.get(url).text
```

```python
soup=BeautifulSoup(response,'lxml')           # stores the HTML page into BeautifulSoup
soup.prettify                                         # prettify method turn BeautifulSoup parse tree into nicely formatted unicode string with a separate line for each tag and each string
```

```python
# extracting year

table = soup.find('table')
year = []
for row in table.find_all('tr')[1:]:
    columns=row.find_all('th')
    year.append([column.text for column in columns[0:2]])
for i in range(0,len(year)):
    if year[i]==[]:
        year[i]=year[i-1]
year
```

```python
# extracting columns

table = soup.find('table')
new_table = []
for row in table.find_all('tr')[1:]:
    columns=row.find_all('td')
    new_table.append([column.text for column in columns[0:2]])
    
new_table
```

```python
final=[]
for a,b in zip(year,new_table):
    c=a+b
    final.append(c)
final
```

```python
df=pd.DataFrame(final,columns=['Year','Title','Role'])
df['Year']=df['Year'].str.replace('\n','')
df['Title']=df['Title'].str.replace('\n','')
df['Role']=df['Role'].str.replace('\n','')
df
```

```python
write=pd.ExcelWriter('Leonardo_det.xlsx')
df.to_excel(write,'film-detail')
write.save()

print('Data has been saved to excel file')
```
